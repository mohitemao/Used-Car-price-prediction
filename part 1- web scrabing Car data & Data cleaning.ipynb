# Library required for Web scrabing

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

# First we will get the link for 50 pages of cartrade website

```python
webpage_links=[]
for i in range(1,51):
    url=f'https://www.cartrade.com/buy-used-cars/mumbai/c/#city=1&sc=-1&so=-1&pn={i}'
    webpage_links.append(url)
```

```python
webpage_links
```

# Creating the new link on each page link

```python
new_link=[]
for i in webpage_links:
    web=requests.get(i).text
    soup=BeautifulSoup(web,'lxml')
    List=soup.find_all('h2',class_='h2heading truncate')
    for x in List:
        page_link='https://www.cartrade.com'+x.a['href']   
        new_link.append(page_link)
```

```python
print(len(new_link))
```

```python
new_link[0:5]
```

# Extract the data from the all link 

```python
detail_list=[]
for i in new_link:
    web2=requests.get(i).text
    soup2=BeautifulSoup(web2,'lxml')
    car_name=soup2.find('h1',class_='h_2').text.split()[1]       # car name
    model_name=soup2.find('h1',class_='h_2').text.split()[2]     # model name
    try:
     price=soup2.find('p',class_='price-value').text.split()[1]  # Price of car
    except:
        pass
    try:
     original_price=soup2.find('p',class_='price-box__value').text.split()[1]     # Original price of car
    except:
        pass
    tr=soup2.find_all('td')

    all_spec=[]
    for i in tr:
        spec=i.text.split()
        all_spec.append(spec)

    # no of owner
    try:
     no_of_owner=all_spec[9][0]  # no of owner of car
    except:
        pass
    try:
     KMS_Driven=all_spec[5][0]
    except:
        pass
    try:
     fuel_type=all_spec[3][0]
    except:
        pass
    try:
     MANUFACTURING_YEAR=all_spec[15][0]
    except:
        pass
    try:
     colour=all_spec[7][0]
    except:
        pass
    try:
     capacity=soup2.find_all('li',class_='specs-details__item')[2]
    except:
        pass
    try:
     seating_capacity=capacity.text.split()[6]
    except:
        pass

    milage=soup2.find_all('span',class_='truncate-text')
    milage_car=[]
    for i in milage:
        milage1=i.text.split()
        milage_car.append(milage1)
    try:
     mil=milage_car[5][0]
    except:
        pass
        
        
    disc={'car_name':car_name,
          'model_name':model_name,
          'no_of_owner':no_of_owner,
          'KMS_Driven':KMS_Driven,
          'fuel_type':fuel_type,
          'MANUFACTURING_YEAR':MANUFACTURING_YEAR,
          'colour':colour,
          'seating_capacity':seating_capacity,
          'mil': mil,
          'price':price,
          'original_price':original_price,}
        
    detail_list.append(disc)
    
    
    
```

```python
detail_list
```

```python
df=pd.DataFrame(detail_list,columns=['car_name','model_name','no_of_owner','KMS_Driven','fuel_type','MANUFACTURING_YEAR','colour','seating_capacity','mil','price','original_price'])
df
```

```python
df.to_csv('Car_data.csv')
```

```python
# Since the number having the commas in values not possible to change the datatype to numerical 
# We will replace the comma with nothing
# e.g.
# number= '10,000'
# new_num=number.replace(',','')
# new_num
```

```python
driven=[]
for i in car_data.KMS_Driven:
    try:
     new=i.replace(',','')         ## We will replace the comma with nothing
    except:
            pass
    driven.append(new)
        
car_data['KMS_Driven']=driven    
```

```python
kms_driven=[]
for i in car_data.KMS_Driven:
    try:
     KMS=int(i)
    except:
        pass
    kms_driven.append(KMS)
car_data['KMS_Driven']=kms_driven    
```

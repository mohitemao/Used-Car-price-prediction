```python
diesel=pd.read_csv('diesel_car.csv')
petrol=pd.read_csv('petrol_car.csv')
cng=pd.read_csv('cng_car.csv')
electric=pd.read_csv('electric_car.csv')
```

```python
petrol_cars=pd.read_csv('petrol_car.csv')
cng_cars=pd.read_csv('cng_car.csv')
electric_cars=pd.read_csv('electric_car.csv')
```

```python
petrol_cars.drop(columns='Unnamed: 0',inplace=True)
cng_cars.drop(columns='Unnamed: 0',inplace=True)
electric_cars.drop(columns='Unnamed: 0', inplace=True)

```

```python


data=pd.concat([petrol_cars,diesel],axis=0)
```

```python
data=pd.concat([data,cng_cars],axis=0)
```

```python
data=pd.concat([data,electric_cars],axis=0)
```

```python
data
```

```python
# how many cars are for sale in all four category
ax=sns.countplot(data=data,x='fuel_type')
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

* we have most petrols cars followed by diesels cas sng and then elctric
* since the elctric and cng cars gave new in market its sale rate may be low as compared to petrol and diesel cars

```python
# top 10 cars which are sailing as per fuel type
plt.figure(figsize=(15,7))
sns.countplot(data=data,x='car_name',order=data.car_name.value_counts().iloc[:10].index,hue='fuel_type')
              
```
